For positive curvature we know that
 $$ ds^2 = dr^2 + R^2 Sin^2(r/R)d\theta^2$$ 
 
I think that this is independent of azimuthal direction.
We can also assume that the object is along a geodesic so that ds = 0.

So we get:

$$ dr^2 = -R^2 Sin^2(r/R)d\theta^2 $$
$$ dr = i R Sin(r/R)d\theta $$
$$ d\theta = dr/(R Sinh((ir)/R) $$

When  $$ r -> \pi R $$
    $$ d\theta = dr/ R Sinh(i*\pi)$$

```python
import numpy as np
import matplotlib.pyplot as plt

#limits 0<=t<=100
a=0
b=100

#steps
N=10000

#initial value a(t_0)= 1
IV =(0.0, 1)

#stepSize
h=(b-a)/N

#ODE
def f(g,t):
    return 70/t**(3/2)

#make arrays to hold values
t = np.arange(a,b+h,h)
w = np.zeros((N+1,))

#Set initival values
t[0],w[0] = IV

#Apply Euler's method
for i in range(1,N+1):
    w[i]=w[i-1]+h*f(t[i-1],w[i-1])
    
#exact solution
def h(t):
    return (175*t-1)**(2/5)

plt.plot (t,w, label='Approximation')
plt.plot(t,h(t))
plt.title("Scaling constant with respect to time Numerical vs Algebraical Approximation")
plt.xlabel('t')
plt.ylabel('a[t]')
plt.show()
```
