```python
class Verse: #Creates a class that will perform calculations based on input Omega parameters
    
        def __init__(self, omega_r, omega_m, omega_lam):
            
            self.H0 = 2.2683e-18 #in seconds
            self.omega_r = omega_r #radiation
            self.omega_m = omega_m #matter
            self.omega_lam = omega_lam #lambda
            self.omega_0 = (self.omega_r + self.omega_m + self.omega_lam)
            
            pass
    
        def Friedmann(self, x):
            t = np.sqrt(1/(self.H0**2.0*(self.omega_r*(x**-2.0) + self.omega_m*(x**-1.0) + self.omega_lam*(x**2.0)
                                             + (1.0 - self.omega_0 + 0.j))))
            if imag(t)==0.j:
                return abs(t)
        
            else:
                pass
```

Problem 2:

Einstein Ring: $\theta_E^{2} = \frac{4 GMD_{LS}}{c^{2}D_{L}D_{S}}$

with isothermal sphere mass density: $\rho(r) = \frac{\sigma_{\nu}^{2}}{2 \pi Gr^{2}}$

To find the mass enclosed within a sphere of radius R

$M(R) = \int \rho(\vec{r}) d \vec{r} = \int_{0}^{R} \int_{0}^{\pi} \int_{0}^{2\pi} \frac{\sigma_{\nu}^{2}}{2 \pi Gr^{2}} r^{2} \sin{\theta} d\phi d\theta dr = \int_{0}^{R} \int_{0}^{\pi} \int_{0}^{2\pi} \frac{\sigma_{\nu}^{2}}{2 \pi G} \sin{\theta} d\phi d\theta dr = 2\pi \int_{0}^{R} \int_{0}^{\pi} \frac{\sigma_{\nu}^{2}}{2 \pi G} \sin{\theta} d\theta dr = 2 \int_{0}^{R} \frac{\sigma_{\nu}^{2}}{G} d\theta dr = \frac{2 \sigma_{\nu}^{2} R}{G}$

To solve for the mass enclised within an Einstein ring 

$M = \frac{\theta_{E}^{2} c^{2} D_{L}D_{S}}{4 G D_{LS}}$

$\rightarrow M = \frac{2 \sigma_{\nu}^{2} R}{G} = \frac{\theta_{E}^{2} c^{2} D_{L}D_{S}}{4 G D_{LS}} \rightarrow \sigma_{\nu}^{2} = \frac{\theta_{E}^{2} c^{2} D_{L}D_{S}}{8 R D_{LS}}$

Applying the small angle approximation $\theta \approx \frac{R}{D_{L}} \rightarrow R = \theta D_{L}$

$\sigma_{\nu}^{2} = \frac{\theta^{2} c^{2} D_{L}D_{S}}{8 R D_{LS}} \rightarrow \sigma_{\nu}^{2} \approx \frac{\theta^{2} c^{2} D_{L}D_{S}}{8 \theta D_{L} D_{LS}} \approx \frac{\theta c^{2} D_{S}}{8 D_{LS}} \rightarrow \sigma_{\nu} \approx c (\frac{\theta}{8} \frac{D_{S}}{D_{LS}})^{1/2}$
