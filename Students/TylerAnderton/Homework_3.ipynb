If we are given the Friedmann Equation 
$(\frac{\dot{a}}{a})^2=\frac{8{\pi}G}{3c^2}\epsilon-\frac{{\kappa}c^2}{R_0^2a^2}$ 
and the Acceleration Equation 
$\frac{\ddot{a}}{a}=-\frac{4{\pi}G}{3c^2}(\epsilon+3P)$,
then we can derive the Fluid Equation 
$\dot{\epsilon}+3\frac{\dot{a}}{a}(\epsilon+P)=0$

We start by multiplying the Freidmann Equation by $a^2$, then we take the time derivative:

$\dot{a}^2=\frac{8{\pi}G}{3c^2}{\epsilon}a^2-\frac{{\kappa}c^2}{R_0^2}$

$2\dot{a}\ddot{a}=\frac{8{\pi}G}{3c^2}(\dot{\epsilon}a^2+2{\epsilon}a\dot{a})$

Now we divide by $2a\dot{a}$ and see that our result is equal to the Acceleration Equation:

$\frac{\ddot{a}}{a}=\frac{4{\pi}G}{3c^2}(\dot{\epsilon}\frac{a}{\dot{a}}+2\epsilon)=-\frac{4{\pi}G}{3c^2}(\epsilon+3P)$

By cancelling like terms and simple rearrangement, the familiar version of the Fluid Equation appears:

$\dot{\epsilon}\frac{a}{\dot{a}}+2\epsilon=-\epsilon-3P$

$\dot{\epsilon}+3\frac{\dot{a}}{a}(\epsilon+P)=0$

For the first part of this problem, I found the comoving distance as a function of redshift of the form 

$r(z)=\int_0^z\frac{cdz'}{H(z')}$

```python
%matplotlib inline

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import scipy.special

#define constants
H0 = 70.0
c = 3.0*(10**5)

#define H
def H(H0, omegaR, omegaM, omegaL, z):
    omega0 = omegaR + omegaM + omegaL
    return H0 * (omegaR * (1+z)**4 + omegaM * (1+z)**3 + omegaL + (1-omega0) * (1+z)**2)**(0.5)

#define integrand
integrand = lambda z: c/H(H0, omegaR,omegaM,omegaL,z)

#define age (nah)
age = lambda z: 980./((1 + z) * H(H0, omegaR,omegaM,omegaL,z))

#establish arrays for redshift, comoving distance, and age
steps = 51
redshift = np.linspace(0., 10., steps)
rA = np.zeros(steps)
ageA = np.zeros(steps)
rB = np.zeros(steps)
ageB = np.zeros(steps)
ageR = np.zeros(steps)

#establish omega values for case a: flat universe
omegaR = 0.0
omegaM = 0.3
omegaL = 0.7

#evaluate rA
for i,z in enumerate(redshift):
#    print i,z, integrate.quad(integrand, 0, z)
    rA[i] = integrate.quad(integrand, 0, z)[0]    
    
#evaluate age of the universe
ageU = integrate.quad(age, 0, 100000)[0]
    
#evaluate ageA
for i,z in enumerate(redshift):
    ageA[i] = ageU-integrate.quad(age, 0, z)[0]
    
#testing so far  
#print redshift
#print rA
#print ageA
print "Age of the universe: " +str(ageU)+ "Gyr"

#establish omega values for case b: closed universe
omegaR = 0.0
omegaM = 10.
omegaL = 0.

#evaluate rB
for i,z in enumerate(redshift):
#    print i,z, integrate.quad(integrand, 0, z)
    rB[i] = integrate.quad(integrand, 0, z)[0]
    
#evaluate ageB
for i,z in enumerate(redshift):
    ageB[i] = ageU-integrate.quad(age, 0, z)[0]
    
#test again    
#print rB
#print ageB

#plotting comoving distance
flat, = plt.plot(redshift, rA, label = 'Flat: $\Omega_{r,0}=0$, $\Omega_{m,0}=0.3$, $\Omega_{\Lambda,0}=0.7$')
closed, = plt.plot(redshift, rB, label = 'Closed: $\Omega_{r,0}=0$, $\Omega_{m,0}=10$, $\Omega_{\Lambda,0}=0$')
plt.legend(handles=[flat, closed], loc='best')
plt.xlabel('Redshift $z$')
plt.ylabel('Comoving Distance $r$ (Mpc)')
plt.title('Comoving Distance vs Redshift for Flat and Closed Universes')
plt.show()
```

For the second part of this problem, I first calculated the "lookback time" from the equation

$t=\int_0^z\frac{dz'}{(1+z)H(z')}$

I then used this equation to find the age of the universe by calculating the lookback time at high $z=10000$. At this magnitude of redshift, the lookback time is fairly constant, indicating we can use this value as a good estimate of the actual age of the universe:

$t_{total}=13.5$ Gyr

Then using this time as the standard, I found the age vs redshift by subtracting the lookback time from the total age of the universe:

$t_{age}=t_{total}-\int_0^z\frac{dz'}{(1+z)H(z')}$

I then plotted the ages versus redshift below along with the radiation only universe curve in problem 3.
