## [ ONE ] 

We used three equations when deriving the Friedman equations:
\begin{equation} 
    \frac{\ddot{a(t)}}{a(t)} = -\frac{4\pi G}{3 c^{2}}[\epsilon(t) + 3P(t)]
\end{equation}

\begin{equation}
    \left ( \frac{\dot{a(t)}}{a(t)} \right )^{2} = \frac{8 \pi G \epsilon(t)}{3c^{2}} - \frac{kc^{2}}{a(t)^{2}}
\end{equation}

and 

\begin{equation}
    \dot{\epsilon(t)} + \frac{3\dot{a(t)}}{a(t)}(\epsilon + P) = 0.
\end{equation}

In fact, only two of these are independent. Show how we get the 3rd of these equations from the first two.

From the above equations, the first equation can be derived from the last two. Starting with the second equation, the Friedman equation, it can be rearranged by multiplying each side by $a^{2}$:

\begin{equation}
    \dot{a(t)}^{2} = \frac{8\pi G}{3c^{2}}a(t)\epsilon(t) - \frac{kc^{2}}{R_{0}^{2}}
\end{equation}
I will drop the explicit time dependence from here on out. Taking the time derivative on each side yields:

\begin{equation}
    2\dot{a}\ddot{a} = \frac{8\pi G}{3c^{2}}(2a\dot{a}\epsilon + a^{2}\dot{\epsilon})
\end{equation}
Divide each side by $2a\dot{a}$:

\begin{equation}
    \frac{\ddot{a}}{a} = \frac{4\pi G}{3c^{2}}(2\epsilon + \frac{a}{\dot{a}}\dot{\epsilon})
\end{equation}
Now, introducing the third equation from above, the Fluid equation, can be rearranged by moving $\dot{\epsilon}$ to the opposite side, and mulitply each side by $-\frac{a}{\dot{a}}$:

\begin{equation}
    \dot{\epsilon}\frac{a}{\dot{a}} = -3(\epsilon + P)
\end{equation}
This can then be substituted into the equation we left behind in the step before last, and simplified to the final form:

\begin{equation}
    \frac{\ddot{a}}{a} = -\frac{4\pi G}{3c^{2}}(\epsilon + 3P) \ \Box
\end{equation}
The acceleration equation.

Using your understanding of how to integrate (from the second homework) write a program that integrates the Friedman equation,

\begin{equation}
    \left ( \frac{H(z)}{H_{0}} \right ) ^{2} = \left ( \Omega_{r,0}(1+z)^{4} + \Omega_{m,0}(1+z)^{3} + \Omega_{\Lambda,0} + (1-\Omega_{0})(1+z)^{2} \right )
\end{equation}

Using this program, calculate and plot (a) the comoving distance from z=0 to 10, (b) the age of the universe (at a given redshift) from z=0 to z=10. Calculate these results for the two universes given below.

(a) a flat universe with $\Omega_{m,0} = 0.3, \Omega_{r,0} = 0, \Omega_{\Lambda,0} = 0.7$

(b) a closed universe with $\Omega_{m,0}= 10, \Omega_{r,0} = 0, \Omega_{\Lambda} = 0$

Note: to calculate the age of the universe remember that H(z) is directly related to dz/dt. Use $H_{0} = 70\ km s^{-1} Mpc^{-1}$ in your calculation and plot the age in Gyr.

(a) The equation for comoving distance is:

\begin{equation}
    r = c \int_{0}^{z} \frac{dz}{H(z)}
\end{equation}

Solving for H(z) in the first equation and plugging it into the comoving equation yields:

\begin{equation}
    r = \frac{c}{H_{0}} \int_{0}^{z} \frac{dz}{\sqrt{\Omega_{r,0}(1+z)^{4} + \Omega_{m,0}(1+z)^{3} + \Omega_{\Lambda,0} + (1-\Omega_{0})(1+z)^{2}}}
\end{equation}

This can then be integrated from 0 to 10 to give the comoving distance, given the appropriate variables.

(b) Starting with the first equation, replacing z with $\frac{1}{a} - 1$, and simplifying, it becomes:

\begin{equation}
    \frac{H(a)^{2}}{H_{0}^{2}} = \frac{\Omega_{r,0}}{a^{4}} + \frac{\Omega_{m,0}}{a^{3}} + \Omega_{\Lambda,0} + \frac{(1 - \Omega_{0})}{a^{2}}
\end{equation}

This can be turn into an integral with the relationship: $H = \frac{\dot{a}}{a}$

Plug this in, multiply each side by $a^{2}$, take the square root, invert the whole thing, multiply each side by $dt$, and integrate, I get:

\begin{equation}
    t = \frac{1}{H_{0}} \int_{0}^{a} \frac{da}{\sqrt{\frac{\Omega_{r,0}}{a^{2}} + \frac{\Omega_{m,0}}{a} + \Omega_{\Lambda, 0}a^{2} + (1-\Omega_{0})}}
\end{equation}

Since $a = \frac{1}{1+z}$, when $z=10$, $a=\frac{1}{11}$. The above integral can then be calculated, given the appropriate variables, to find the age of the universe at the time when the scale factor equals 1/11, or when z equals 10. It can be integrated from 0 to 1 to give the total age of the universe. 

#### Okay, time to calculate these integrals for each of the two cases above.

(a) a flat universe with $\Omega_{m,0} = 0.3, \Omega_{r,0} = 0, \Omega_{\Lambda,0} = 0.7$

Equation for Comoving distance:
\begin{equation}
    r = \frac{c}{H_{0}} \int_{0}^{10}  \frac{dz}{\sqrt{(0.3(1+z)^{3} + 0.7)}}
\end{equation}

Equation for age of the universe at a = 1/11:
\begin{equation}
    t = \frac{1}{H_{0}} \int_{0}^{\frac{1}{11}} \frac{da}{\sqrt{\frac{0.3}{a} + 0.7a^{2}}}
\end{equation}

Both of these equations are integrated in the cell below.

```python
from scipy import integrate
from decimal import Decimal
import numpy as np
import matplotlib.pyplot as plt

#Setting my variables
c = 3.0*(10**8)
z_stop = Decimal(10)
scale = np.linspace(0, 1.0/11.00, 100)
H = 70 * (3.241*(10**-20)) #Change units if needed. 

#Define the comoving equation and age of the universe equation.
def comoving(z):
    return (0.3*(1+z)**3 + 0.7)**(-0.5)
def age(a):
    return (0.3/a + 0.7*(a**2))**(-0.5)

#Getting the solution to comoving distance.
sol = integrate.quad(comoving,  0, float(z_stop))
dist_com = c * 1/H * sol[0]

#Getting the solution to the age of the universe.
sol_2 = integrate.quad(age, 0, 1.0/11.00)
univ_age = sol_2[0] * (1/H) * (1/(3.154*10**7)) / (10**9)

#Print my results, change some units
print 'Comoving distance = ', dist_com * (3.241*(10**(-23))), 'Mpc'
print 'Age of the Universe at z=10 = ', univ_age, 'Gyrs'

#Plot the scale factor versus the age of the universe:
plt.plot(scale, age(scale) * (1/H) * (1/(3.154*10**7)) / (10**9))
plt.xlabel('Scale Factor a(t)')
plt.ylabel('Age of the Universe [Gyrs]')
plt.title('Scale Factor vs. Age of the Universe (a)')
```

```python
from scipy import integrate
from decimal import Decimal
import numpy as np
import matplotlib.pyplot as plt

#Setting my variables
c = 3.0*(10**8)
z_stop = Decimal(10)
scale = np.linspace(0, 1.0/11.00, 100)
H = 70 * (3.241*(10**-20)) #Change units if needed.

#Define the comoving equation and age of the universe equation.
def comoving_b(z):
    return (10*(1+z)**(3) -9.0*(1+z)**(2))**(-0.5)
def age_b(a):
    return ((10/a) - 9)**(-0.5)

#Getting the solution to comoving distance.
sol_3 = integrate.quad(comoving_b,  0, float(z_stop))
dist_com = c * 1/H * sol_3[0]

#Getting the solution to the age of the universe.
sol_4 = integrate.quad(age_b, 0, 1.0/11.0)
z_age = sol_4[0] * (1/H) * (1/(3.154*10**7)) / (10**9)

#Print my results, change some units
print 'Comoving distance = ', dist_com * (3.241*(10**(-23))), 'Mpc'
print 'Age of the Universe at z=10 = ', str(z_age) , 'Gyrs'

#Plot the scale factor versus the age of the universe:
plt.plot(scale, age_b(scale) * (1/H) * (1/(3.154*10**7)) / (10**9))
plt.xlabel('Scale Factor a(t)')
plt.ylabel('Age of the Universe [Gyrs]')
plt.title('Scale Factor vs. Age of the Universe (b)')
```

## [ THREE ] 

Use the program to find the value of $\Omega_{r}$ for a radiation only open universe (no cosmological constant) that would give the same age of the universe (for the current day) as our benchmark flat universe with $\Omega_{m,0} = 0.3, \Omega_{\Lambda,0} = 0.7$.

The equation for the total age of the universe is as follows:

\begin{equation}
    t = \frac{1}{H_{0}} \int_{0}^{1} \frac{da}{\sqrt{\frac{\Omega_{r,0}}{a^{2}} + \frac{\Omega_{m,0}}{a} + \Omega_{\Lambda, 0}a^{2} + (1-\Omega_{0})}}
\end{equation}

For a radiation only, open universe, this equation becomes:

\begin{equation}
    t = \frac{1}{H_{0}} \int_{0}^{1} \frac{da}{\sqrt{\frac{\Omega_{r,0}}{a^{2}}  + (1-\Omega_{r,0})}}
\end{equation}

I am looking for a value for $\Omega_{r,0}$ which makes this integral equal to approximately 13.47 Gyrs (calculated at the top of the cell below for reference). Then, by adding steps to the value of $\Omega_{r,0}$ as the age of the universe is calculated and compared to the benchmark value at every step through the use of a simple while statement, I reach a value of $\Omega_{r,0}$ = 0.001387. 


```python
from scipy import integrate
from scipy import optimize as opt
from scipy.optimize import minimize
from decimal import Decimal
import numpy as np
import matplotlib.pyplot as plt

#Setting my variables
H = 70 * (3.241*(10**-20)) #Change units if needed.
scale = np.linspace(0, 1, 100)
#Start at zero for my initial guess of Omega_r,0
r = 0
step = 0.000001

#Benchmark model age of the universe 
def age_bench(a):
    return (0.3/(a) + 0.7*(a**2))**(-0.5)
bench = integrate.quad(age_bench, 0, 1)
bench_age = bench[0] * (1/H) * (1/(3.154*10**7)) / (10**9)
print 'Age of Benchmark Universe today = ', bench_age, ' Gyrs'

#Define the age of the universe equation for radiation only universe.
def age(a):
    return (r/(a**2) - r + 1)**(-0.5)

#Get the solution to the age of radiation only universe 
sol = integrate.quad(age, 0, 1)
new_age = sol[0] * (1/H) * (1/(3.154*10**7)) / (10**9)

#While age of the radiation only universe is greater than the age of the benchmark universe, add steps to the r value and recalulate the age of the universe, i.e. optimize for r. 
while (new_age >= bench_age):
    r += step
    sol = integrate.quad(age, 0, 1)
    new_age = sol[0] * (1/H) * (1/(3.154*10**7)) / (10**9)

print 'Age of a radiation only, open universe = ', sol[0] * (1/H) * (1/(3.154*10**7)) / (10**9), 'Gyrs, with an \Omega_{r,0} value = ', r

#Plot the scale factor versus the age of the universe:
plt.plot(scale, age_bench(scale) * (1/H) * (1/(3.154*10**7)) / (10**9),'-', label = 'Benchmark Model')
plt.plot(scale, age(scale) * (1/H) * (1/(3.154*10**7)) / (10**9), label = r'Closed, rad. only: $\Omega_{r,0}$ = ' + str(r))
plt.xlabel('Scale Factor a(t)')
plt.ylabel('Age of the Universe [Gyrs]')
plt.legend(loc = 4)
plt.title('Age of the Universe vs. Scale Factor')
```
