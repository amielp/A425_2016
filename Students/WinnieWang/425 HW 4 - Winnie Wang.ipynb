##### Problem 1: Redshift of Two Distant Galaxies
Recall that the redshift, "z", can be represented in this form: $1 + z = \frac{a(t_o)}{a(t_e)}$, or equivalent to $z = \frac{a(t_o)}{a(t_e)} - 1$.
##### Part A:
In the observer's perspective, $a(t_o) = 1$, so the redshift of the first and second galaxy can be denoted as: $1 + z_1 = \frac{1}{a(t_e,1)}$ and $1 + z_2 = \frac{1}{a(t_e,2)}$ respectively. 

Using the same expression, the redshift of galaxy 2 observed from galaxy 1 can be written this way: $1 + z_{2,1} = \frac{a(t_{o,1})}{a(t_{e,2})}$, where $z_{2,1}$ is the redshift of 2 observed from 1. 

Since we are observing in the frame of galaxy 1, $a(t_{o,1}) = a(t_{e,1})$. Hence the redshift observed from galaxy 1 is: $1 + z_{2,1} = \frac{a(t_{e,1})}{a(t_{e,2})}$. 

Replacing $a(t_{e,1})$ and $a(t_{e,2})$ with the first expressions result in:

$$1 + z_{2,1} = \frac{a(t_{e,1})}{a(t_{e,2})} = \frac{\frac{1}{1 + z_1}}{\frac{1}{1 + z_2}} = \frac{1 + z_2}{1 + z_1}$$

Rearranging this above expression shows that the redshift of galaxy 2 observed from galaxy 1 shows that:

$$z_{2,1} = \frac{1 + z_2}{1 + z_1} - 1$$

##### Part B:
To find the angular distance there are two relations that can be used. Firstly, in a flat universe, the angular and proper distance are related by: 

$$d_A = \frac{d_p (t_o)}{(1 + z)}$$

Secondly, the general expression relating the state parameter, $\omega$, and proper distance is:

$$d_p(t_o) = \frac{2c}{H_o(1+3\omega)}[1 - {(1 + z)}^{-\frac{1 + 3\omega}{2}}]$$

In a matter-only universe, $\omega = 0$. Putting this in to the above expression leaves the proper distance relation as:

$$d_p(t_o) = \frac{2c}{H_o}[1 - {(1 + z)}^{-\frac{1}{2}}] = \frac{2c}{H_o}[1 - \frac{1}{\sqrt{1+z}}]$$

Putting this expression into $d_A$, it is observed that:

$$d_A = \frac{d_p (t_o)}{(1 + z)} \rightarrow \frac{2c}{H_o(1 + z)}[1 - \frac{1}{\sqrt{1 + z}}]$$

Since c = speed of light = $3*10^8 m/s$ = $3*10^5 km/s$ and $H_o = 70 km/(Mpc\bullet s)$, $\frac{2c}{H_o}$ evaluates to ${(6*{10}^{5})}/{70}$ Mpc, or about $8571.428571$ Mpc. Putting this number in makes the angular distance:

$$d_A = \frac{d_p (t_o)}{(1 + z)} \rightarrow \frac{8571.428571}{(1 + z)}[1 - \frac{1}{\sqrt{1 + z}}]$$

Putting $z_1 = 0.5$ and $z_2 = 1.6$, it is observed that putting these numbers into the expression solved in part A, the redshift of galaxy 2 observed from galaxy 1 shows that:

$$z_{2,1} = \frac{1 + 1.6}{1 + 0.5} - 1 \approx 1.733$$

Hence, the angular distances of based on $z_1$, $z_2$, and $z_{2,1}$ are:

$$d_{A, z_1} = \frac{8571.428571}{(1 + 0.5)}[1 - \frac{1}{\sqrt{1 + 0.5}}] \approx 1048.6\,Mpc = 1.0486\,Gpc$$

$$d_{A, z_2} = \frac{8571.428571}{(1 + 1.6)}[1 - \frac{1}{\sqrt{1 + 1.6}}] \approx 1252.2\,Mpc = 1.2522\,Gpc$$

$$d_{A, z_{2,1}} = \frac{8571.428571}{(1 + 1.733)}[1 - \frac{1}{\sqrt{1 + 1.733}}] \approx 1239.2\,Mpc = 1.2392\,Gpc$$

##### Problem 2: (Corrected) Ryden 8.5 with N = 1000 galaxies in the half- mass sphere, and $r_h = 1.5\,Mpc$.

##### Mean number density of galaxies within the half-mass radius?
The density of galaxies, "n", in a given volume is the total number of galaxies contained per volume. To put this in variables, this is equivalent to: $n  = \frac{N}{V_{sphere}}$.

Recall that the volume of sphere is: $V_{sphere} = \frac{4}{3}\pi {r}^{3}$, or specifically $V_{sphere} = \frac{4}{3}\pi {r_h}^{3}$ for a sphere with the half-mass radius. 

Consequently, a sphere with a half-mass radius has a volume of: $V_{sphere} = \frac{4}{3} (3.1415) {1.5\,Mpc}^{3} \approx 14.137\,{Mpc}^{3}$

Hence the mean number of density of galaxies that are within this sphere of a half-mass radius is:


$$ n = \frac{N}{V_{sphere}} \rightarrow n = \frac{1000\,galaxies}{14.137\,{Mpc}^{3}} \approx 70.737\,\frac{galaxies}{{Mpc}^{3}}$$

##### Mean collision distance?
For a galaxy to collide with another galaxy, the mean free path of the galaxy must be calculated. 

Specifically, we want that the product of the cross section area, $\sigma_{galaxy} = {10}^{-3}\,Mpc$, travel path, $d_{path}$, and the number of galaxies to be 1:

$$\sigma_{galaxy} d_{path} n_{galaxy} = 1\,galaxy$$

In solving for the mean free path, rearranging this expression results:

$$ d_{path} = \frac{1\,galaxy}{n_{galaxy} \sigma_{galaxy}}$$

Plugging the known numbers from the first part results to:

$$ d_{path} = \frac{1\,galaxy}{70.737\frac{galaxies}{{Mpc}^{3}}{{10}^{-3}{Mpc}^{2}}} \approx 14.137\,Mpc$$


##### Average collision time? Greater/smaller than Hubble time?
From the general relationship between distance, velocity and time, or in particular, $d = vt$, it is determined that the average time for collision is obtained from dividing the average distance by the average velocity, or equivalently, 

$$<t> = \frac{<d>}{<v>}$$

However, since we do not know any galaxy's average velocity in the cluster, the relationship between the average velocity squared and velocity dispersion can be used. The relationship specifically shows that: 

$$ <{v}^{2}> = 3{\sigma}^{2} \rightarrow <v> = \sqrt{3{\sigma}^{2}} $$

Given that the velocity dispersion in the Coma cluster is $\sigma \approx 880 km*{s}^{-1}$, and that the mean collision distance, $<d> \approx 14.137\,Mpc$:

$$ <t> = \frac{<d>}{<v>} = \frac{<d>}{<\sqrt{3{\sigma}^{2}}>}\,\approx\,\frac{14.137\,Mpc}{\sqrt{3}*880\frac{km}{s}} = \frac{4.362*{10}^{20}\,km}{1524.2047...\frac{km}{s}} \approx 2.86*{10}^{17}\,s \approx 9.07\,Gyr$$

This number is not larger than the Hubble time, which is roughly 14.4 $Gyr$, so it should be expected galaxies in the Coma cluster had collided, if the Coma cluster is older than 9.07 $Gyr$.

##### Compare with collision time of Andromeda and Milky Way to galaxies in Coma cluster?

Since we are observing the collision of Andromeda in our own galaxy, we are observing in a frame that is not moving. Given that we know the average velocity of travel, it can be estimated that the mean collision time between our galaxy and Andromeda is:

$$ <t>\,= \frac{<d>}{<v>} = \frac{700kpc}{120\frac{km}{s}} = \frac{2.16*{10}^{19}km}{120\frac{km}{s}} = 1.8*{10}^{17}\,seconds \approx 5.708\,Gyr$$
