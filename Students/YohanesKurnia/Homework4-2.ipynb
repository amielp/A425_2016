```python
#Gravitational lensing simulation using M ~ 1*10^11 M_sun Distance to source ~1Gpc
import numpy as np
import matplotlib.pyplot as py
from matplotlib import colors as c

#Setting colormaps for the source and lensed object.
cMap = c.ListedColormap(['w','g'])
cMap2 = c.ListedColormap(['None','r'])

#Setting constants
massOfSun = 2*10**30
numberOfStars = 1.00*10**11
gravConst = 6.67*10**-11
speedOfLight = 3*10**8
D_ds = 1000*3.086*10**22

resolutionSize = 1001
canvasCenter = resolutionSize/2-0.5

theta_e_rad = np.sqrt(4*gravConst*numberOfStars*massOfSun/(speedOfLight)**2 * (0.5)/(D_ds*0.5))
theta_e = theta_e_rad*canvasCenter/(6*10**-6)

#Drawing einstein ring as a 
einsteinRing = py.Circle((canvasCenter,canvasCenter),theta_e,linestyle ='--',fill='')


#Setting the source plane and the lens plane.
sourcePlane = np.zeros((resolutionSize,resolutionSize))
lensPlane = np.zeros((resolutionSize,resolutionSize))

#Center of the circular source with radius of about 0.3µrad. 
ci,cj= canvasCenter,canvasCenter
cr= 0.2/6*resolutionSize


#Setting the source object's size
I,J = np.meshgrid(np.arange(sourcePlane.shape[0]),np.arange(sourcePlane.shape[1]))
beta = np.sqrt((I-ci)**2+(J-cj)**2)
sourcePlane[np.where(beta<cr)]=1

#Uses for loop to iterate through source plane to look for any 1's  
#and if value is 1, get the value of beta by getting the distance from
#canvasCenter to [i,j], and gamma the angle beta makes with the axis. 
#Then use lens equation to get the value of theta. Then use theta and gamma
#To find the change in x and y and from the center of lensPlane add x and y and
#change value to 1. This will be the lensed image.
for i in range(len(sourcePlane)):
    for j in range(len(sourcePlane)):
        if sourcePlane[i][j] == 1:
            if i != canvasCenter:
                beta = np.sqrt((i-canvasCenter)**2+(j-canvasCenter)**2)
                gamma = np.arctan((j-canvasCenter)/(i-canvasCenter)) 
            else:
                beta = 0
                gamma = 0
            theta_1 = 1/2*(beta + np.sqrt(beta**2 + 4*theta_e**2))
            theta_2 = 1/2*(beta - np.sqrt(beta**2 + 4*theta_e**2))

            x_val_lens_1 = theta_1*np.cos(gamma)
            y_val_lens_1 = theta_1*np.sin(gamma)
            lensPlane[canvasCenter+x_val_lens_1][canvasCenter+y_val_lens_1]=1
            x_val_lens_2 = theta_2*np.cos(gamma)
            y_val_lens_2 = theta_2*np.sin(gamma)
            lensPlane[canvasCenter+x_val_lens_2][canvasCenter+y_val_lens_2]=1

            
#FIGURE CONFIGURATIONS
fig = py.figure()
ax1= fig.add_subplot(111)
ax1.add_artist(einsteinRing)
ax1.pcolormesh(sourcePlane,cmap = cMap,label="Source Mass")
ax1.pcolormesh(lensPlane,cmap=cMap2,label="Lensed Image")
ax1.set_aspect('equal')
ax1.set_xlim((0,resolutionSize))
ax1.set_ylim((0,resolutionSize))
ticks = np.arange(0, len(lensPlane), (resolutionSize-1)/6)
labels = range(-6,ticks.size,2)
py.yticks(ticks,labels)
py.xticks(ticks, labels)
py.xlabel(r"$\theta_x [\mu rad]$")
py.ylabel(r"$\theta_y [\mu rad]$")
py.grid()
py.show()

```

```python
from PIL import Image as image
from PIL import ImageEnhance as IEnh
import numpy as np
import matplotlib.pyplot as py
from IPython.display import Image as Img
from IPython.display import display as Display



#Replace the source object with the HST image. then the image would be lensed.

#Setting constants
massOfSun = 2*10**30
numberOfStars = 1.00*10**11
gravConst = 6.67*10**-11
speedOfLight = 3*10**8
D_ds = 1000*3.086*10**22

resolutionSize = 1001
canvasCenter = resolutionSize/2-0.5

im =  image.open("pic.jpg")
col,row =  im.size

theta_e_rad = np.sqrt(4*gravConst*numberOfStars*massOfSun/(speedOfLight)**2 * (0.5)/(D_ds*0.5))
theta_e = theta_e_rad*canvasCenter/(6*10**-6)

#Drawing einstein ring as a 
einsteinRing = py.Circle((canvasCenter,canvasCenter),theta_e,linestyle ='--',fill='')


#Setting the source plane and the lens plane.
sourcePlane = np.zeros((resolutionSize,resolutionSize))
lensPlane = np.zeros((resolutionSize,resolutionSize),dtype='object')

#Center of the circular source with radius of about 0.3µrad. 
ci,cj= canvasCenter,canvasCenter
cr= 1/6*resolutionSize

#Setting the source object's size
I,J = np.meshgrid(np.arange(sourcePlane.shape[0]),np.arange(sourcePlane.shape[1]))
beta = np.sqrt((I-ci)**2+(J-cj)**2)
sourcePlane[np.where(beta<cr)]=1


#Uses for loop to iterate through source plane to look for any 1's  
#and if value is 1, get the value of beta by getting the distance from
#canvasCenter to [i,j], and gamma the angle beta makes with the axis. 
#Then use lens equation to get the value of theta. Then use theta and gamma
#To find the change in x and y and from the center of lensPlane add x and y and
#change value to 1. This will be the lensed image.
for i in range(len(sourcePlane)):
    for j in range(len(sourcePlane)):
        if sourcePlane[i][j] == 1:
            if i != canvasCenter:
                beta = np.sqrt((i-canvasCenter)**2+(j-canvasCenter)**2)
                gamma = np.arctan((j-canvasCenter)/(i-canvasCenter)) 
            else:
                beta = 0
                gamma = 0
            theta_1 = 1/2*(beta + np.sqrt(beta**2 + 4*theta_e**2))
            theta_2 = 1/2*(beta - np.sqrt(beta**2 + 4*theta_e**2))

            x_val_lens_1 = theta_1*np.cos(gamma)
            y_val_lens_1 = theta_1*np.sin(gamma)
            lensPlane[canvasCenter+x_val_lens_1][canvasCenter+y_val_lens_1]=im.getpixel((i,j))
            x_val_lens_2 = theta_2*np.cos(gamma)
            y_val_lens_2 = theta_2*np.sin(gamma)
            lensPlane[canvasCenter+x_val_lens_2][canvasCenter+y_val_lens_2]=im.getpixel((i,j))


lensedTuple = [tuple(pixels) for pixels in lensPlane]
# print(lensedTuple)
lensedImage = image.new(im.mode,im.size)
# print(type(test1image))
for i in range(len(lensedTuple)):
    for j in range(len(lensedTuple)):
        lensedImage.putpixel((i,j),lensedTuple[i][j])

finalImage = image.blend(im,lensedImage,0.5)
enhancer = IEnh.Brightness(finalImage)
finalImage = enhancer.enhance(3.0)
im.save('original.jpg')
finalImage.save('lensed.jpg')

Original = Img(filename='original.jpg')
Lensed = Img(filename='lensed.jpg')


#First picture is original and the second picture is lensed image.
Display(Original,Lensed)





```
